```python
import matplotlib.pyplot as plt
import numpy as np
#from sympy import *
#init_printing()
```

# Pair-wise potentials

```python
r_0 = 2 # Å
```

```python
D_e = 661.4 # kJ/mol
```

```python
r = np.arange(0.1,5,0.005)
```

## Lennard-Jones

```python
V_LJ = D_e*((r_0/r)**12-2*(r_0/r)**6)
```

```python
plt.plot(r, V_LJ, linewidth=2)
plt.axhline(0, color="k", linewidth=1)
plt.axvline(2, color="k", linewidth=1)
plt.xlim(0,5)
plt.ylim(-800,400)
plt.xlabel(r"$r$ [Å]")
plt.ylabel(r"$E$ [kJ/mol]")
plt.show()
```

## Morse

```python
a_M = 3
```

```python
#V_M = D_e*(1 - np.exp(-a_M*(r-r_0)))**2 - D_e
V_M = D_e*(np.exp(-2*a_M*(r-r_0))-2*np.exp(-a_M*(r-r_0)))
```

```python
plt.plot(r, V_M, linewidth=2)
plt.axhline(0, color="k", linewidth=1)
plt.axvline(2, color="k", linewidth=1)
plt.xlim(0,5)
plt.ylim(-800,400)
plt.xlabel(r"$r$ [Å]")
plt.ylabel(r"$E$ [kJ/mol]")
plt.show()
```

## Coulomb–Pauli

```python
a_CP = 10
```

```python
b_CP = 100
```

```python
l = 1389 # kJ/mol
```

```python
q = 1
```

```python
V_CP = -l*q**2/r*(1-b_CP*np.exp(-a_CP*(r+0.765-r_0)))
```

```python
plt.plot(r, V_CP, linewidth=2)
plt.axhline(0, color="k", linewidth=1)
plt.axvline(2, color="k", linewidth=1)
plt.xlim(0,5)
plt.ylim(-800,400)
plt.xlabel(r"$r$ [Å]")
plt.ylabel(r"$E$ [kJ/mol]")
plt.set_major_locator(MultipleLocator(1))
plt.show()
```

## Harmonic oscillator

```python
k = 13000
```

```python
V_HO = 1/2*k*(r-r_0)**2 - D_e
```

```python
plt.plot(r, V_HO, linewidth=2)
plt.axhline(0, color="k", linewidth=1)
plt.axvline(2, color="k", linewidth=1)
plt.xlim(0,5)
plt.ylim(-800,400)
plt.xlabel(r"$r$ [Å]")
plt.ylabel(r"$E$ [kJ/mol]")
plt.show()
```

## Universal binding curve

```python
a_U = 7500
```

```python
b_U = a_U/np.exp(1)/D_e  #D_e = a/(b*e)
```

```python
V_U = -a_U*(r+1/b_U-r_0)*np.exp(-b_U*(r+1/b_U-r_0))
```

```python
plt.plot(r, V_U, linewidth=2)
plt.axhline(0, color="k", linewidth=1)
plt.axvline(2, color="k", linewidth=1)
plt.xlim(0,5)
plt.ylim(-800,400)
plt.xlabel(r"$r$ [Å]")
plt.ylabel(r"$E$ [kJ/mol]")
plt.show()
```

```python
plt.plot(r, V_HO, "-", linewidth=2, label="Harmonic oscillator")
plt.plot(r, V_M, "-.", linewidth=2, label="Morse")
plt.plot(r, V_U, "--", linewidth=2, label="Universal") 
plt.plot(r, V_LJ, ":", linewidth=2, label="Lennard-Jones")
plt.plot(r, V_CP, "-", linewidth=2, label="Coulomb−Pauli")
plt.xlim(1.5,3.5)
plt.ylim(-700,500)
plt.xlabel(r"$r$ [Å]")
plt.ylabel(r"$E$ [kJ/mol]")
plt.legend(loc="upper right")
plt.show()
```
